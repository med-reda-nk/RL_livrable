```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import time

class GridWorldEnv(gym.Env):
    def __init__(self, size=5):
        super().__init__()
        self.size = size
        self.observation_space = spaces.Box(low=0, high=size-1, shape=(2,), dtype=np.int32)
        self.action_space = spaces.Discrete(4)  # 0: haut, 1: bas, 2: gauche, 3: droite

        self.agent_pos = np.array([0, 0])
        self.goal_pos = np.array([size-1, size-1])

    def reset(self, seed=None):
        super().reset(seed=seed)
        self.agent_pos = np.array([0, 0])
        return self._get_obs(), {}

    def _get_obs(self):
        return self.agent_pos.copy()

    def _pos_to_state(self, pos):
        return pos[1] * self.size + pos[0]

    def step(self, action):
        moves = {
            0: np.array([0, -1]),  # haut
            1: np.array([0, 1]),   # bas
            2: np.array([-1, 0]),  # gauche
            3: np.array([1, 0])    # droite
        }

        new_pos = self.agent_pos + moves[action]

        if 0 <= new_pos[0] < self.size and 0 <= new_pos[1] < self.size:
            self.agent_pos = new_pos

        done = np.array_equal(self.agent_pos, self.goal_pos)
        reward = 10.0 if done else -0.1

        return self._get_obs(), reward, done, False, {}

# ===== AGENT ALÉATOIRE =====
class RandomAgent:
    def __init__(self, env):
        self.env = env
        self.name = "Random"

    def select_action(self, state):
        return self.env.action_space.sample()

    def train(self, num_episodes):
        pass  # Pas d'entraînement

# ===== VALUE ITERATION =====
class ValueIterationAgent:
    def __init__(self, env, gamma=0.99, theta=1e-6):
        self.env = env
        self.gamma = gamma
        self.theta = theta
        self.name = "Value Iteration"
        self.V = np.zeros(env.size * env.size)
        self.policy = np.zeros(env.size * env.size, dtype=int)

    def train(self, num_episodes=None):
        while True:
            delta = 0
            for s in range(self.env.size * self.env.size):
                v = self.V[s]
                x, y = s % self.env.size, s // self.env.size

                if [x, y] == [self.env.size-1, self.env.size-1]:
                    self.V[s] = 0
                    continue

                action_values = []
                for a in range(4):
                    self.env.agent_pos = np.array([x, y])
                    _, reward, done, _, _ = self.env.step(a)
                    next_s = self.env._pos_to_state(self.env.agent_pos)
                    action_values.append(reward + self.gamma * self.V[next_s])

                self.V[s] = max(action_values)
                self.policy[s] = np.argmax(action_values)
                delta = max(delta, abs(v - self.V[s]))

            if delta < self.theta:
                break

    def select_action(self, state):
        s = self.env._pos_to_state(state)
        return self.policy[s]

# ===== POLICY ITERATION =====
class PolicyIterationAgent:
    def __init__(self, env, gamma=0.99, theta=1e-6):
        self.env = env
        self.gamma = gamma
        self.theta = theta
        self.name = "Policy Iteration"
        self.V = np.zeros(env.size * env.size)
        self.policy = np.random.randint(0, 4, env.size * env.size)

    def policy_evaluation(self):
        while True:
            delta = 0
            for s in range(self.env.size * self.env.size):
                v = self.V[s]
                x, y = s % self.env.size, s // self.env.size

                if [x, y] == [self.env.size-1, self.env.size-1]:
                    self.V[s] = 0
                    continue

                action = self.policy[s]
                self.env.agent_pos = np.array([x, y])
                _, reward, done, _, _ = self.env.step(action)
                next_s = self.env._pos_to_state(self.env.agent_pos)
                self.V[s] = reward + self.gamma * self.V[next_s]
                delta = max(delta, abs(v - self.V[s]))

            if delta < self.theta:
                break

    def policy_improvement(self):
        policy_stable = True
        for s in range(self.env.size * self.env.size):
            x, y = s % self.env.size, s // self.env.size

            if [x, y] == [self.env.size-1, self.env.size-1]:
                continue

            old_action = self.policy[s]
            action_values = []
            for a in range(4):
                self.env.agent_pos = np.array([x, y])
                _, reward, done, _, _ = self.env.step(a)
                next_s = self.env._pos_to_state(self.env.agent_pos)
                action_values.append(reward + self.gamma * self.V[next_s])

            self.policy[s] = np.argmax(action_values)
            if old_action != self.policy[s]:
                policy_stable = False

        return policy_stable

    def train(self, num_episodes=None):
        while True:
            self.policy_evaluation()
            if self.policy_improvement():
                break

    def select_action(self, state):
        s = self.env._pos_to_state(state)
        return self.policy[s]

# ===== MONTE CARLO =====
class MonteCarloAgent:
    def __init__(self, env, gamma=0.99, epsilon=0.1):
        self.env = env
        self.gamma = gamma
        self.epsilon = epsilon
        self.name = "Monte Carlo"
        self.Q = defaultdict(lambda: np.zeros(env.action_space.n))
        self.returns = defaultdict(list)

    def select_action(self, state, training=True):
        s = tuple(state)
        if training and np.random.random() < self.epsilon:
            return self.env.action_space.sample()
        return np.argmax(self.Q[s])

    def train(self, num_episodes):
        for episode in range(num_episodes):
            trajectory = []
            state, _ = self.env.reset()
            done = False

            while not done:
                action = self.select_action(state, training=True)
                next_state, reward, done, _, _ = self.env.step(action)
                trajectory.append((tuple(state), action, reward))
                state = next_state

            G = 0
            visited = set()
            for t in reversed(range(len(trajectory))):
                state, action, reward = trajectory[t]
                G = reward + self.gamma * G

                if (state, action) not in visited:
                    visited.add((state, action))
                    self.returns[(state, action)].append(G)
                    self.Q[state][action] = np.mean(self.returns[(state, action)])

# ===== Q-LEARNING =====
class QLearningAgent:
    def __init__(self, env, alpha=0.1, gamma=0.99, epsilon=0.1):
        self.env = env
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.name = "Q-Learning"
        self.Q = defaultdict(lambda: np.zeros(env.action_space.n))

    def select_action(self, state, training=True):
        s = tuple(state)
        if training and np.random.random() < self.epsilon:
            return self.env.action_space.sample()
        return np.argmax(self.Q[s])

    def train(self, num_episodes):
        for episode in range(num_episodes):
            state, _ = self.env.reset()
            done = False

            while not done:
                action = self.select_action(state, training=True)
                next_state, reward, done, _, _ = self.env.step(action)

                s = tuple(state)
                next_s = tuple(next_state)

                best_next_action = np.argmax(self.Q[next_s])
                td_target = reward + self.gamma * self.Q[next_s][best_next_action]
                td_error = td_target - self.Q[s][action]
                self.Q[s][action] += self.alpha * td_error

                state = next_state

# ===== ÉVALUATION DES AGENTS =====
def evaluate_agent(agent, env, num_episodes=100):
    total_rewards = []
    total_steps = []

    for _ in range(num_episodes):
        state, _ = env.reset()
        done = False
        episode_reward = 0
        steps = 0

        while not done and steps < 200:
            action = agent.select_action(state)
            state, reward, done, _, _ = env.step(action)
            episode_reward += reward
            steps += 1

        total_rewards.append(episode_reward)
        total_steps.append(steps)

    return np.mean(total_rewards), np.std(total_rewards), np.mean(total_steps), np.std(total_steps)

# ===== PROGRAMME PRINCIPAL =====
def main():
    env = GridWorldEnv(size=5)

    # Créer les agents
    agents = [
        RandomAgent(env),
        ValueIterationAgent(env),
        PolicyIterationAgent(env),
        MonteCarloAgent(env),
        QLearningAgent(env)
    ]

    results = {}

    print("\n" + "="*60)
    print("ENTRAÎNEMENT ET ÉVALUATION DES AGENTS")
    print("="*60 + "\n")

    for agent in agents:
        print(f" Agent: {agent.name}")

        # Entraînement
        start_time = time.time()
        if agent.name == "Random":
            print("   Pas d'entraînement nécessaire...")
        elif agent.name in ["Monte Carlo", "Q-Learning"]:
            print("   Entraînement en cours (1000 épisodes)...")
            agent.train(num_episodes=1000)
        else:
            print("   Entraînement en cours...")
            agent.train()
        training_time = time.time() - start_time

        # Évaluation
        print("   Évaluation en cours (100 épisodes)...")
        mean_reward, std_reward, mean_steps, std_steps = evaluate_agent(agent, env, num_episodes=100)

        results[agent.name] = {
            'mean_reward': mean_reward,
            'std_reward': std_reward,
            'mean_steps': mean_steps,
            'std_steps': std_steps,
            'training_time': training_time
        }

        print(f"     Récompense moyenne: {mean_reward:.2f} ± {std_reward:.2f}")
        print(f"     Étapes moyennes: {mean_steps:.2f} ± {std_steps:.2f}")
        print(f"     Temps d'entraînement: {training_time:.2f}s\n")

    # Visualisation
    print("Génération des graphiques...\n")

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    agent_names = list(results.keys())
    rewards = [results[name]['mean_reward'] for name in agent_names]
    reward_stds = [results[name]['std_reward'] for name in agent_names]
    steps = [results[name]['mean_steps'] for name in agent_names]
    step_stds = [results[name]['std_steps'] for name in agent_names]
    times = [results[name]['training_time'] for name in agent_names]

    colors = ['red', 'blue', 'green', 'orange', 'purple']

    # Graphique 1: Récompenses moyennes
    axes[0, 0].bar(agent_names, rewards, yerr=reward_stds, color=colors, alpha=0.7, capsize=5)
    axes[0, 0].set_ylabel('Récompense Moyenne', fontsize=11)
    axes[0, 0].set_title('Performance: Récompenses', fontsize=12, fontweight='bold')
    axes[0, 0].grid(axis='y', alpha=0.3)
    axes[0, 0].tick_params(axis='x', rotation=15)

    # Graphique 2: Nombre d'étapes
    axes[0, 1].bar(agent_names, steps, yerr=step_stds, color=colors, alpha=0.7, capsize=5)
    axes[0, 1].set_ylabel('Nombre d\'étapes moyen', fontsize=11)
    axes[0, 1].set_title('Efficacité: Étapes jusqu\'à l\'objectif', fontsize=12, fontweight='bold')
    axes[0, 1].grid(axis='y', alpha=0.3)
    axes[0, 1].tick_params(axis='x', rotation=15)

    # Graphique 3: Temps d'entraînement
    axes[1, 0].bar(agent_names, times, color=colors, alpha=0.7)
    axes[1, 0].set_ylabel('Temps (secondes)', fontsize=11)
    axes[1, 0].set_title('Temps d\'entraînement', fontsize=12, fontweight='bold')
    axes[1, 0].grid(axis='y', alpha=0.3)
    axes[1, 0].tick_params(axis='x', rotation=15)

    # Graphique 4: Tableau récapitulatif
    axes[1, 1].axis('off')
    table_data = []
    for name in agent_names:
        table_data.append([
            name,
            f"{results[name]['mean_reward']:.2f}",
            f"{results[name]['mean_steps']:.1f}",
            f"{results[name]['training_time']:.2f}s"
        ])

    table = axes[1, 1].table(cellText=table_data,
                            colLabels=['Agent', 'Récompense', 'Étapes', 'Temps'],
                            cellLoc='center',
                            loc='center',
                            colWidths=[0.3, 0.2, 0.2, 0.2])
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 2)

    for i in range(len(agent_names) + 1):
        if i == 0:
            table[(i, 0)].set_facecolor('#40466e')
            table[(i, 1)].set_facecolor('#40466e')
            table[(i, 2)].set_facecolor('#40466e')
            table[(i, 3)].set_facecolor('#40466e')
            table[(i, 0)].set_text_props(weight='bold', color='white')
            table[(i, 1)].set_text_props(weight='bold', color='white')
            table[(i, 2)].set_text_props(weight='bold', color='white')
            table[(i, 3)].set_text_props(weight='bold', color='white')
        else:
            table[(i, 0)].set_facecolor(colors[i-1])
            table[(i, 0)].set_alpha(0.3)

    plt.suptitle('Comparaison des Agents de Reinforcement Learning',
                fontsize=16, fontweight='bold', y=0.995)
    plt.tight_layout()
    plt.show()

    print("="*60)
    print("✅ ÉVALUATION TERMINÉE")
    print("="*60)

if __name__ == "__main__":
    main()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import time
import os

class GridWorldEnv(gym.Env):
    def __init__(self, size=5):
        super().__init__()
        self.size = size
        self.observation_space = spaces.Box(low=0, high=size-1, shape=(2,), dtype=np.int32)
        self.action_space = spaces.Discrete(4)  # 0: haut, 1: bas, 2: gauche, 3: droite

        self.agent_pos = np.array([0, 0])
        self.goal_pos = np.array([size-1, size-1])

    def reset(self, seed=None):
        super().reset(seed=seed)
        self.agent_pos = np.array([0, 0])
        return self._get_obs(), {}

    def _get_obs(self):
        return self.agent_pos.copy()

    def _pos_to_state(self, pos):
        return pos[1] * self.size + pos[0]

    def step(self, action):
        moves = {
            0: np.array([0, -1]),  # haut
            1: np.array([0, 1]),   # bas
            2: np.array([-1, 0]),  # gauche
            3: np.array([1, 0])    # droite
        }

        new_pos = self.agent_pos + moves[action]

        if 0 <= new_pos[0] < self.size and 0 <= new_pos[1] < self.size:
            self.agent_pos = new_pos

        done = np.array_equal(self.agent_pos, self.goal_pos)
        reward = 10.0 if done else -0.1

        return self._get_obs(), reward, done, False, {}

    def render(self):
        grid = [['.' for _ in range(self.size)] for _ in range(self.size)]
        grid[self.goal_pos[1]][self.goal_pos[0]] = 'G'
        grid[self.agent_pos[1]][self.agent_pos[0]] = 'A'
        for row in grid:
            print(' '.join(row))
        print()

# ===== AGENT ALÉATOIRE =====
class RandomAgent:
    def __init__(self, env):
        self.env = env
        self.name = "Random"

    def select_action(self, state):
        return self.env.action_space.sample()

    def train(self, num_episodes):
        pass  # Pas d'entraînement

# ===== VALUE ITERATION =====
class ValueIterationAgent:
    def __init__(self, env, gamma=0.99, theta=1e-6):
        self.env = env
        self.gamma = gamma
        self.theta = theta
        self.name = "Value Iteration"
        self.V = np.zeros(env.size * env.size)
        self.policy = np.zeros(env.size * env.size, dtype=int)

    def train(self, num_episodes=None):
        while True:
            delta = 0
            for s in range(self.env.size * self.env.size):
                v = self.V[s]
                x, y = s % self.env.size, s // self.env.size

                if [x, y] == [self.env.size-1, self.env.size-1]:
                    self.V[s] = 0
                    continue

                action_values = []
                for a in range(4):
                    self.env.agent_pos = np.array([x, y])
                    _, reward, done, _, _ = self.env.step(a)
                    next_s = self.env._pos_to_state(self.env.agent_pos)
                    action_values.append(reward + self.gamma * self.V[next_s])

                self.V[s] = max(action_values)
                self.policy[s] = np.argmax(action_values)
                delta = max(delta, abs(v - self.V[s]))

            if delta < self.theta:
                break

    def select_action(self, state):
        s = self.env._pos_to_state(state)
        return self.policy[s]

# ===== POLICY ITERATION =====
class PolicyIterationAgent:
    def __init__(self, env, gamma=0.99, theta=1e-6):
        self.env = env
        self.gamma = gamma
        self.theta = theta
        self.name = "Policy Iteration"
        self.V = np.zeros(env.size * env.size)
        self.policy = np.random.randint(0, 4, env.size * env.size)

    def policy_evaluation(self):
        while True:
            delta = 0
            for s in range(self.env.size * self.env.size):
                v = self.V[s]
                x, y = s % self.env.size, s // self.env.size

                if [x, y] == [self.env.size-1, self.env.size-1]:
                    self.V[s] = 0
                    continue

                action = self.policy[s]
                self.env.agent_pos = np.array([x, y])
                _, reward, done, _, _ = self.env.step(action)
                next_s = self.env._pos_to_state(self.env.agent_pos)
                self.V[s] = reward + self.gamma * self.V[next_s]
                delta = max(delta, abs(v - self.V[s]))

            if delta < self.theta:
                break

    def policy_improvement(self):
        policy_stable = True
        for s in range(self.env.size * self.env.size):
            x, y = s % self.env.size, s // self.env.size

            if [x, y] == [self.env.size-1, self.env.size-1]:
                continue

            old_action = self.policy[s]
            action_values = []
            for a in range(4):
                self.env.agent_pos = np.array([x, y])
                _, reward, done, _, _ = self.env.step(a)
                next_s = self.env._pos_to_state(self.env.agent_pos)
                action_values.append(reward + self.gamma * self.V[next_s])

            self.policy[s] = np.argmax(action_values)
            if old_action != self.policy[s]:
                policy_stable = False

        return policy_stable

    def train(self, num_episodes=None):
        while True:
            self.policy_evaluation()
            if self.policy_improvement():
                break

    def select_action(self, state):
        s = self.env._pos_to_state(state)
        return self.policy[s]

# ===== MONTE CARLO =====
class MonteCarloAgent:
    def __init__(self, env, gamma=0.99, epsilon=0.1):
        self.env = env
        self.gamma = gamma
        self.epsilon = epsilon
        self.name = "Monte Carlo"
        self.Q = defaultdict(lambda: np.zeros(env.action_space.n))
        self.returns = defaultdict(list)

    def select_action(self, state, training=False):
        s = tuple(state)
        if training and np.random.random() < self.epsilon:
            return self.env.action_space.sample()
        return np.argmax(self.Q[s])

    def train(self, num_episodes):
        for episode in range(num_episodes):
            trajectory = []
            state, _ = self.env.reset()
            done = False

            while not done:
                action = self.select_action(state, training=True)
                next_state, reward, done, _, _ = self.env.step(action)
                trajectory.append((tuple(state), action, reward))
                state = next_state

            G = 0
            visited = set()
            for t in reversed(range(len(trajectory))):
                state, action, reward = trajectory[t]
                G = reward + self.gamma * G

                if (state, action) not in visited:
                    visited.add((state, action))
                    self.returns[(state, action)].append(G)
                    self.Q[state][action] = np.mean(self.returns[(state, action)])

# ===== Q-LEARNING =====
class QLearningAgent:
    def __init__(self, env, alpha=0.1, gamma=0.99, epsilon=0.1):
        self.env = env
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.name = "Q-Learning"
        self.Q = defaultdict(lambda: np.zeros(env.action_space.n))

    def select_action(self, state, training=False):
        s = tuple(state)
        if training and np.random.random() < self.epsilon:
            return self.env.action_space.sample()
        return np.argmax(self.Q[s])

    def train(self, num_episodes):
        for episode in range(num_episodes):
            state, _ = self.env.reset()
            done = False

            while not done:
                action = self.select_action(state, training=True)
                next_state, reward, done, _, _ = self.env.step(action)

                s = tuple(state)
                next_s = tuple(next_state)

                best_next_action = np.argmax(self.Q[next_s])
                td_target = reward + self.gamma * self.Q[next_s][best_next_action]
                td_error = td_target - self.Q[s][action]
                self.Q[s][action] += self.alpha * td_error

                state = next_state

# ===== ÉVALUATION DES AGENTS =====
def evaluate_agent(agent, env, num_episodes=100):
    total_rewards = []
    total_steps = []

    for _ in range(num_episodes):
        state, _ = env.reset()
        done = False
        episode_reward = 0
        steps = 0

        while not done and steps < 200:
            action = agent.select_action(state)
            state, reward, done, _, _ = env.step(action)
            episode_reward += reward
            steps += 1

        total_rewards.append(episode_reward)
        total_steps.append(steps)

    return np.mean(total_rewards), np.std(total_rewards), np.mean(total_steps), np.std(total_steps)

# ===== SIMULATION STEP-BY-STEP =====
def simulate_episode(agent, env):
    state, _ = env.reset()
    done = False
    steps = 0
    total_reward = 0.0
    action_names = {0: "haut", 1: "bas", 2: "gauche", 3: "droite"}

    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"Simulation pour l'agent: {agent.name}")
    print("État initial:")
    env.render()
    time.sleep(1)

    while not done and steps < 200:
        action = agent.select_action(state)
        state, reward, done, _, _ = env.step(action)
        total_reward += reward
        steps += 1

        os.system('cls' if os.name == 'nt' else 'clear')
        print(f"Simulation pour l'agent: {agent.name}")
        print(f"Étape: {steps}")
        print(f"Action prise: {action_names[action]}")
        print(f"Récompense: {reward}")
        print(f"Récompense totale: {total_reward}")
        env.render()
        time.sleep(1)

    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"Simulation pour l'agent: {agent.name}")
    print("État final:")
    env.render()
    print(f"Total des étapes: {steps}")
    print(f"Récompense totale: {total_reward}")
    if done:
        print("Objectif atteint!")
    else:
        print("Limite d'étapes atteinte.")

# ===== PROGRAMME PRINCIPAL =====
def main():
    env = GridWorldEnv(size=5)

    # Créer les agents
    agents = [
        RandomAgent(env),
        ValueIterationAgent(env),
        PolicyIterationAgent(env),
        MonteCarloAgent(env),
        QLearningAgent(env)
    ]

    results = {}

    print("\n" + "="*60)
    print("ENTRAÎNEMENT ET ÉVALUATION DES AGENTS")
    print("="*60 + "\n")

    for agent in agents:
        print(f"🤖 Agent: {agent.name}")

        # Entraînement
        start_time = time.time()
        if agent.name == "Random":
            print("   Pas d'entraînement nécessaire...")
        elif agent.name in ["Monte Carlo", "Q-Learning"]:
            print("   Entraînement en cours (1000 épisodes)...")
            agent.train(num_episodes=1000)
        else:
            print("   Entraînement en cours...")
            agent.train()
        training_time = time.time() - start_time

        # Évaluation
        print("   Évaluation en cours (100 épisodes)...")
        mean_reward, std_reward, mean_steps, std_steps = evaluate_agent(agent, env, num_episodes=100)

        results[agent.name] = {
            'mean_reward': mean_reward,
            'std_reward': std_reward,
            'mean_steps': mean_steps,
            'std_steps': std_steps,
            'training_time': training_time
        }

        print(f"   ✅ Récompense moyenne: {mean_reward:.2f} ± {std_reward:.2f}")
        print(f"   📊 Étapes moyennes: {mean_steps:.2f} ± {std_steps:.2f}")
        print(f"   ⏱️  Temps d'entraînement: {training_time:.2f}s\n")

    # Visualisation
    print("Génération des graphiques...\n")

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    agent_names = list(results.keys())
    rewards = [results[name]['mean_reward'] for name in agent_names]
    reward_stds = [results[name]['std_reward'] for name in agent_names]
    steps = [results[name]['mean_steps'] for name in agent_names]
    step_stds = [results[name]['std_steps'] for name in agent_names]
    times = [results[name]['training_time'] for name in agent_names]

    colors = ['red', 'blue', 'green', 'orange', 'purple']

    # Graphique 1: Récompenses moyennes
    axes[0, 0].bar(agent_names, rewards, yerr=reward_stds, color=colors, alpha=0.7, capsize=5)
    axes[0, 0].set_ylabel('Récompense Moyenne', fontsize=11)
    axes[0, 0].set_title('Performance: Récompenses', fontsize=12, fontweight='bold')
    axes[0, 0].grid(axis='y', alpha=0.3)
    axes[0, 0].tick_params(axis='x', rotation=15)

    # Graphique 2: Nombre d'étapes
    axes[0, 1].bar(agent_names, steps, yerr=step_stds, color=colors, alpha=0.7, capsize=5)
    axes[0, 1].set_ylabel('Nombre d\'étapes moyen', fontsize=11)
    axes[0, 1].set_title('Efficacité: Étapes jusqu\'à l\'objectif', fontsize=12, fontweight='bold')
    axes[0, 1].grid(axis='y', alpha=0.3)
    axes[0, 1].tick_params(axis='x', rotation=15)

    # Graphique 3: Temps d'entraînement
    axes[1, 0].bar(agent_names, times, color=colors, alpha=0.7)
    axes[1, 0].set_ylabel('Temps (secondes)', fontsize=11)
    axes[1, 0].set_title('Temps d\'entraînement', fontsize=12, fontweight='bold')
    axes[1, 0].grid(axis='y', alpha=0.3)
    axes[1, 0].tick_params(axis='x', rotation=15)

    # Graphique 4: Tableau récapitulatif
    axes[1, 1].axis('off')
    table_data = []
    for name in agent_names:
        table_data.append([
            name,
            f"{results[name]['mean_reward']:.2f}",
            f"{results[name]['mean_steps']:.1f}",
            f"{results[name]['training_time']:.2f}s"
        ])

    table = axes[1, 1].table(cellText=table_data,
                            colLabels=['Agent', 'Récompense', 'Étapes', 'Temps'],
                            cellLoc='center',
                            loc='center',
                            colWidths=[0.3, 0.2, 0.2, 0.2])
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 2)

    for i in range(len(agent_names) + 1):
        if i == 0:
            table[(i, 0)].set_facecolor('#40466e')
            table[(i, 1)].set_facecolor('#40466e')
            table[(i, 2)].set_facecolor('#40466e')
            table[(i, 3)].set_facecolor('#40466e')
            table[(i, 0)].set_text_props(weight='bold', color='white')
            table[(i, 1)].set_text_props(weight='bold', color='white')
            table[(i, 2)].set_text_props(weight='bold', color='white')
            table[(i, 3)].set_text_props(weight='bold', color='white')
        else:
            table[(i, 0)].set_facecolor(colors[i-1])
            table[(i, 0)].set_alpha(0.3)

    plt.suptitle('Comparaison des Agents de Reinforcement Learning',
                fontsize=16, fontweight='bold', y=0.995)
    plt.tight_layout()
    plt.show()

    print("="*60)
    print("✅ ÉVALUATION TERMINÉE")
    print("="*60)

    # Choix de l'agent pour la simulation
    print("\nChoisissez un agent pour simuler le mouvement étape par étape:")
    for i, agent in enumerate(agents, 1):
        print(f"{i}: {agent.name}")

    while True:
        try:
            choice = int(input("Entrez le numéro de l'agent: "))
            if 1 <= choice <= len(agents):
                selected_agent = agents[choice - 1]
                break
            else:
                print("Choix invalide. Veuillez entrer un numéro valide.")
        except ValueError:
            print("Veuillez entrer un numéro entier.")

    simulate_episode(selected_agent, env)

if __name__ == "__main__":
    main()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
from typing import Optional, Tuple, List
from collections import defaultdict

class MultiGoalGridWorld(gym.Env):

    metadata = {"render_modes": ["human", "rgb_array"], "render_fps": 4}

    def __init__(self, grid_size: int = 10, target_goal: int = 0,
                 num_goals: int = 3, num_obstacles: int = 10):
        super().__init__()

        self.grid_size = grid_size
        self.num_goals = num_goals
        self.target_goal = target_goal
        self.num_obstacles = num_obstacles

        # Define action and observation space
        # Actions: 0=UP, 1=RIGHT, 2=DOWN, 3=LEFT
        self.action_space = spaces.Discrete(4)

        # Observation: agent position (x, y)
        self.observation_space = spaces.Box(
            low=0, high=grid_size-1, shape=(2,), dtype=np.int32
        )

        # Initialize positions
        self.agent_pos = np.array([0, 0])
        self.goals = []
        self.obstacles = []
        self._generate_world()

    def _generate_world(self):
        """Generate random goal and obstacle positions"""
        self.goals = []
        self.obstacles = []

        # Get all available positions (excluding agent start)
        available_positions = [
            (x, y) for x in range(self.grid_size)
            for y in range(self.grid_size)
            if (x, y) != tuple(self.agent_pos)
        ]

        # Total positions needed
        total_needed = min(self.num_goals + self.num_obstacles, len(available_positions))

        # Randomly select positions
        selected_indices = np.random.choice(
            len(available_positions),
            size=total_needed,
            replace=False
        )

        # Assign goals
        for i in range(min(self.num_goals, total_needed)):
            idx = selected_indices[i]
            self.goals.append(np.array(available_positions[idx]))

        # Assign obstacles
        for i in range(self.num_goals, total_needed):
            idx = selected_indices[i]
            self.obstacles.append(np.array(available_positions[idx]))

    def _is_obstacle(self, pos: np.ndarray) -> bool:
        """Check if position contains an obstacle"""
        for obs in self.obstacles:
            if np.array_equal(pos, obs):
                return True
        return False

    def reset(self, seed: Optional[int] = None, options: Optional[dict] = None):
        """Reset the environment"""
        super().reset(seed=seed)

        # Reset agent to fixed start position
        self.agent_pos = np.array([0, 0])

        # Ensure target_goal is valid
        if self.target_goal >= len(self.goals):
            self.target_goal = 0

        return self.agent_pos.copy(), {}

    def step(self, action: int):
        """Execute one step in the environment"""
        # Map actions to movements
        direction_map = {
            0: np.array([-1, 0]),  # UP
            1: np.array([0, 1]),   # RIGHT
            2: np.array([1, 0]),   # DOWN
            3: np.array([0, -1])   # LEFT
        }

        # Calculate new position
        new_pos = self.agent_pos + direction_map[action]

        # Check boundaries
        new_pos = np.clip(new_pos, 0, self.grid_size - 1)

        # Check if new position is an obstacle
        if not self._is_obstacle(new_pos):
            self.agent_pos = new_pos
        # If obstacle, agent stays in current position (collision)

        # Check if reached target goal
        target_pos = self.goals[self.target_goal]
        reached_goal = np.array_equal(self.agent_pos, target_pos)

        # Calculate reward
        if reached_goal:
            reward = 100.0
            terminated = True
        else:
            reward = -1.0  # Step penalty
            terminated = False

        truncated = False

        return self.agent_pos.copy(), reward, terminated, truncated, {}

    def render_grid(self, ax=None, agent_path=None):
        """Render the grid using matplotlib"""
        if ax is None:
            fig, ax = plt.subplots(figsize=(8, 8))

        ax.clear()
        ax.set_xlim(-0.5, self.grid_size - 0.5)
        ax.set_ylim(-0.5, self.grid_size - 0.5)
        ax.set_aspect('equal')
        ax.grid(True, alpha=0.3)
        ax.set_xticks(range(self.grid_size))
        ax.set_yticks(range(self.grid_size))
        ax.invert_yaxis()

        # Draw obstacles
        for obs in self.obstacles:
            rect = Rectangle((obs[1] - 0.5, obs[0] - 0.5), 1, 1,
                           facecolor='black', edgecolor='black')
            ax.add_patch(rect)

        # Draw goals
        for idx, goal in enumerate(self.goals):
            color = 'gold' if idx == self.target_goal else 'lightgray'
            circle = Circle((goal[1], goal[0]), 0.3, color=color, ec='black', linewidth=2)
            ax.add_patch(circle)
            ax.text(goal[1], goal[0], str(idx), ha='center', va='center',
                   fontsize=12, fontweight='bold')

        # Draw agent path if provided
        if agent_path and len(agent_path) > 1:
            path_array = np.array(agent_path)
            ax.plot(path_array[:, 1], path_array[:, 0], 'b--', alpha=0.5, linewidth=2, label='Path')

        # Draw agent
        agent_circle = Circle((self.agent_pos[1], self.agent_pos[0]), 0.3,
                             color='lime', ec='darkgreen', linewidth=2)
        ax.add_patch(agent_circle)

        ax.set_title(f'GridWorld - Target Goal: {self.target_goal}', fontsize=14, fontweight='bold')
        ax.legend(loc='upper right')

        return ax


class QLearningAgent:
    """Q-Learning agent for GridWorld"""

    def __init__(self, action_space_size: int, learning_rate: float = 0.1,
                 discount_factor: float = 0.95, epsilon: float = 1.0,
                 epsilon_decay: float = 0.995, epsilon_min: float = 0.01):
        self.action_space_size = action_space_size
        self.lr = learning_rate
        self.gamma = discount_factor
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min

        # Q-table: defaultdict for dynamic state-action pairs
        self.q_table = defaultdict(lambda: np.zeros(action_space_size))

    def get_action(self, state: np.ndarray) -> int:
        """Choose action using epsilon-greedy policy"""
        state_key = tuple(state)

        if np.random.random() < self.epsilon:
            return np.random.randint(self.action_space_size)  # Explore
        else:
            return np.argmax(self.q_table[state_key])  # Exploit

    def update(self, state: np.ndarray, action: int, reward: float,
               next_state: np.ndarray, done: bool):
        """Update Q-table using Q-learning update rule"""
        state_key = tuple(state)
        next_state_key = tuple(next_state)

        # Q-learning update
        current_q = self.q_table[state_key][action]

        if done:
            target_q = reward
        else:
            max_next_q = np.max(self.q_table[next_state_key])
            target_q = reward + self.gamma * max_next_q

        # Update Q-value
        self.q_table[state_key][action] = current_q + self.lr * (target_q - current_q)

    def decay_epsilon(self):
        """Decay exploration rate"""
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)


def train_and_visualize(grid_size=10, target_goal=0, num_goals=3,
                       num_obstacles=15, episodes=500):
    """Train Q-learning agent and visualize progress"""

    # Create environment
    env = MultiGoalGridWorld(
        grid_size=grid_size,
        target_goal=target_goal,
        num_goals=num_goals,
        num_obstacles=num_obstacles
    )

    # Create agent
    agent = QLearningAgent(
        action_space_size=env.action_space.n,
        learning_rate=0.1,
        discount_factor=0.95,
        epsilon=1.0,
        epsilon_decay=0.995,
        epsilon_min=0.01
    )

    # Training metrics
    episode_rewards = []
    episode_steps = []

    print("Training Q-Learning Agent...")
    print(f"Grid Size: {grid_size}x{grid_size}")
    print(f"Target Goal: {target_goal}")
    print(f"Number of Goals: {num_goals}")
    print(f"Number of Obstacles: {num_obstacles}")
    print(f"Episodes: {episodes}\n")

    # Training loop
    for episode in range(episodes):
        state, _ = env.reset()
        total_reward = 0
        steps = 0
        done = False

        while not done and steps < 200:  # Max steps per episode
            action = agent.get_action(state)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            agent.update(state, action, reward, next_state, done)

            state = next_state
            total_reward += reward
            steps += 1

        agent.decay_epsilon()
        episode_rewards.append(total_reward)
        episode_steps.append(steps)

        if (episode + 1) % 100 == 0:
            avg_reward = np.mean(episode_rewards[-100:])
            avg_steps = np.mean(episode_steps[-100:])
            print(f"Episode {episode + 1}/{episodes} - "
                  f"Avg Reward: {avg_reward:.2f}, "
                  f"Avg Steps: {avg_steps:.2f}, "
                  f"Epsilon: {agent.epsilon:.3f}")

    print("\n✓ Training Complete!\n")

    # Test the trained agent with visualization
    print("Testing Trained Agent (Step-by-Step Visualization)...")
    state, _ = env.reset()
    done = False
    steps = 0
    agent_path = [state.copy()]

    # Create figure for step-by-step visualization
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.flatten()

    # Store snapshots at key intervals
    snapshots = []
    snapshot_steps = []

    while not done and steps < 200:
        action = agent.get_action(state)
        agent.epsilon = 0  # Use greedy policy for testing

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        agent_path.append(next_state.copy())
        state = next_state
        steps += 1

        # Save snapshots at regular intervals or at goal
        if steps == 1 or steps % max(1, steps // 5) == 0 or done:
            snapshots.append((steps, agent_path.copy()))
            snapshot_steps.append(steps)

    # Ensure we show the final state
    if not done or snapshots[-1][0] != steps:
        snapshots.append((steps, agent_path.copy()))
        snapshot_steps.append(steps)

    # Plot snapshots
    for idx in range(min(6, len(snapshots))):
        ax = axes[idx]
        snapshot_idx = int(idx * (len(snapshots) - 1) / 5) if len(snapshots) > 1 else 0
        step_num, path = snapshots[snapshot_idx]

        # Temporarily set agent position to show snapshot
        env.agent_pos = np.array(path[-1])
        env.render_grid(ax, agent_path=path)
        ax.set_title(f'Step {step_num}', fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.savefig('gridworld_qlearning_steps.png', dpi=150, bbox_inches='tight')
    print(f"✓ Step-by-step visualization saved as 'gridworld_qlearning_steps.png'")
    plt.show()

    # Plot training metrics
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot rewards
    ax1.plot(episode_rewards, alpha=0.3, color='blue')
    window = 50
    if len(episode_rewards) >= window:
        moving_avg = np.convolve(episode_rewards, np.ones(window)/window, mode='valid')
        ax1.plot(range(window-1, len(episode_rewards)), moving_avg,
                color='red', linewidth=2, label=f'{window}-episode moving avg')
    ax1.set_xlabel('Episode', fontsize=12)
    ax1.set_ylabel('Total Reward', fontsize=12)
    ax1.set_title('Training Rewards', fontsize=14, fontweight='bold')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot steps
    ax2.plot(episode_steps, alpha=0.3, color='green')
    if len(episode_steps) >= window:
        moving_avg = np.convolve(episode_steps, np.ones(window)/window, mode='valid')
        ax2.plot(range(window-1, len(episode_steps)), moving_avg,
                color='red', linewidth=2, label=f'{window}-episode moving avg')
    ax2.set_xlabel('Episode', fontsize=12)
    ax2.set_ylabel('Steps to Goal', fontsize=12)
    ax2.set_title('Training Steps', fontsize=14, fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('gridworld_qlearning_training.png', dpi=150, bbox_inches='tight')
    print(f"✓ Training metrics saved as 'gridworld_qlearning_training.png'")
    plt.show()

    print(f"\n{'='*60}")
    print(f"RESULTS:")
    print(f"{'='*60}")
    print(f"Goal reached: {done}")
    print(f"Total steps taken: {steps}")
    print(f"Final position: {state}")
    print(f"Target goal position: {env.goals[target_goal]}")
    print(f"Path length: {len(agent_path)}")
    print(f"{'='*60}\n")

    return env, agent, episode_rewards, episode_steps


# Run the training and visualization
if __name__ == "__main__":
    env, agent, rewards, steps = train_and_visualize(
        grid_size=10,
        target_goal=3,
        num_goals=4,
        num_obstacles=3,
        episodes=500
    )
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
from typing import Optional, Tuple, List
from collections import defaultdict
import matplotlib.animation as animation
from IPython.display import HTML

class MultiGoalGridWorld(gym.Env):

    metadata = {"render_modes": ["human", "rgb_array"], "render_fps": 4}

    def __init__(self, grid_size: int = 10, target_goal: int = 0,
                 num_goals: int = 3, num_obstacles: int = 10):
        super().__init__()

        self.grid_size = grid_size
        self.num_goals = num_goals
        self.target_goal = target_goal
        self.num_obstacles = num_obstacles

        # Define action and observation space
        # Actions: 0=UP, 1=RIGHT, 2=DOWN, 3=LEFT
        self.action_space = spaces.Discrete(4)

        # Observation: agent position (x, y)
        self.observation_space = spaces.Box(
            low=0, high=grid_size-1, shape=(2,), dtype=np.int32
        )

        # Initialize positions
        self.agent_pos = np.array([0, 0])
        self.goals = []
        self.obstacles = []
        self._generate_world()

    def _generate_world(self):
        """Generate random goal and obstacle positions"""
        self.goals = []
        self.obstacles = []

        # Get all available positions (excluding agent start)
        available_positions = [
            (x, y) for x in range(self.grid_size)
            for y in range(self.grid_size)
            if (x, y) != tuple(self.agent_pos)
        ]

        # Total positions needed
        total_needed = min(self.num_goals + self.num_obstacles, len(available_positions))

        # Randomly select positions
        selected_indices = np.random.choice(
            len(available_positions),
            size=total_needed,
            replace=False
        )

        # Assign goals
        for i in range(min(self.num_goals, total_needed)):
            idx = selected_indices[i]
            self.goals.append(np.array(available_positions[idx]))

        # Assign obstacles
        for i in range(self.num_goals, total_needed):
            idx = selected_indices[i]
            self.obstacles.append(np.array(available_positions[idx]))

    def _is_obstacle(self, pos: np.ndarray) -> bool:
        """Check if position contains an obstacle"""
        for obs in self.obstacles:
            if np.array_equal(pos, obs):
                return True
        return False

    def reset(self, seed: Optional[int] = None, options: Optional[dict] = None):
        super().reset(seed=seed)

        # Reset agent to fixed start position
        self.agent_pos = np.array([0, 0])

        # Ensure target_goal is valid
        if self.target_goal >= len(self.goals):
            self.target_goal = 0

        return self.agent_pos.copy(), {}

    def step(self, action: int):
        # Map actions to movements
        direction_map = {
            0: np.array([-1, 0]),  # UP
            1: np.array([0, 1]),   # RIGHT
            2: np.array([1, 0]),   # DOWN
            3: np.array([0, -1])   # LEFT
        }

        # Calculate new position
        new_pos = self.agent_pos + direction_map[action]

        # Check boundaries
        new_pos = np.clip(new_pos, 0, self.grid_size - 1)

        # Check if new position is an obstacle
        if not self._is_obstacle(new_pos):
            self.agent_pos = new_pos
        # If obstacle, agent stays in current position (collision)

        # Check if reached target goal
        target_pos = self.goals[self.target_goal]
        reached_goal = np.array_equal(self.agent_pos, target_pos)

        # Calculate reward
        if reached_goal:
            reward = 100.0
            terminated = True
        else:
            reward = -1.0  # Step penalty
            terminated = False

        truncated = False

        return self.agent_pos.copy(), reward, terminated, truncated, {}

    def render_grid(self, ax=None, agent_path=None, current_step=None):
        """Render the grid using matplotlib"""
        if ax is None:
            fig, ax = plt.subplots(figsize=(8, 8))

        ax.clear()
        ax.set_xlim(-0.5, self.grid_size - 0.5)
        ax.set_ylim(-0.5, self.grid_size - 0.5)
        ax.set_aspect('equal')
        ax.grid(True, alpha=0.3)
        ax.set_xticks(range(self.grid_size))
        ax.set_yticks(range(self.grid_size))
        ax.invert_yaxis()

        # Draw obstacles
        for obs in self.obstacles:
            rect = Rectangle((obs[1] - 0.5, obs[0] - 0.5), 1, 1,
                           facecolor='black', edgecolor='black')
            ax.add_patch(rect)

        # Draw goals
        for idx, goal in enumerate(self.goals):
            color = 'gold' if idx == self.target_goal else 'lightgray'
            circle = Circle((goal[1], goal[0]), 0.3, color=color, ec='black', linewidth=2)
            ax.add_patch(circle)
            ax.text(goal[1], goal[0], str(idx), ha='center', va='center',
                   fontsize=12, fontweight='bold')

        # Draw agent path if provided
        if agent_path and len(agent_path) > 1:
            path_array = np.array(agent_path)
            ax.plot(path_array[:, 1], path_array[:, 0], 'b--', alpha=0.5, linewidth=2, label='Path')

        # Draw agent
        agent_circle = Circle((self.agent_pos[1], self.agent_pos[0]), 0.3,
                             color='lime', ec='darkgreen', linewidth=2)
        ax.add_patch(agent_circle)

        title = f'GW - T Goal: {self.target_goal}'
        if current_step is not None:
            title += f' | Step: {current_step}'
        ax.set_title(title, fontsize=14, fontweight='bold')

        if agent_path and len(agent_path) > 1:
            ax.legend(loc='upper right')

        return ax


class QLearningAgent:
    """Q-Learning agent for GridWorld"""

    def __init__(self, action_space_size: int, learning_rate: float = 0.1,
                 discount_factor: float = 0.95, epsilon: float = 1.0,
                 epsilon_decay: float = 0.995, epsilon_min: float = 0.01):
        self.action_space_size = action_space_size
        self.lr = learning_rate
        self.gamma = discount_factor
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min

        # Q-table: defaultdict for dynamic state-action pairs
        self.q_table = defaultdict(lambda: np.zeros(action_space_size))

    def get_action(self, state: np.ndarray) -> int:
        """Choose action using epsilon-greedy policy"""
        state_key = tuple(state)

        if np.random.random() < self.epsilon:
            return np.random.randint(self.action_space_size)  # Explore
        else:
            return np.argmax(self.q_table[state_key])  # Exploit

    def update(self, state: np.ndarray, action: int, reward: float,
               next_state: np.ndarray, done: bool):
        """Update Q-table using Q-learning update rule"""
        state_key = tuple(state)
        next_state_key = tuple(next_state)

        # Q-learning update
        current_q = self.q_table[state_key][action]

        if done:
            target_q = reward
        else:
            max_next_q = np.max(self.q_table[next_state_key])
            target_q = reward + self.gamma * max_next_q

        # Update Q-value
        self.q_table[state_key][action] = current_q + self.lr * (target_q - current_q)

    def decay_epsilon(self):
        """Decay exploration rate"""
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)


def train_and_visualize(grid_size=10, target_goal=0, num_goals=3,
                       num_obstacles=15, episodes=500):
    """Train Q-learning agent and visualize progress"""

    # Create environment
    env = MultiGoalGridWorld(
        grid_size=grid_size,
        target_goal=target_goal,
        num_goals=num_goals,
        num_obstacles=num_obstacles
    )

    # Create agent
    agent = QLearningAgent(
        action_space_size=env.action_space.n,
        learning_rate=0.1,
        discount_factor=0.95,
        epsilon=1.0,
        epsilon_decay=0.995,
        epsilon_min=0.01
    )

    # Training metrics
    episode_rewards = []
    episode_steps = []

    print("Training Q-Learning Agent...")
    print(f"Grid Size: {grid_size}x{grid_size}")
    print(f"Target Goal: {target_goal}")
    print(f"Number of Goals: {num_goals}")
    print(f"Number of Obstacles: {num_obstacles}")
    print(f"Episodes: {episodes}\n")

    # Training loop
    for episode in range(episodes):
        state, _ = env.reset()
        total_reward = 0
        steps = 0
        done = False

        while not done and steps < 200:  # Max steps per episode
            action = agent.get_action(state)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            agent.update(state, action, reward, next_state, done)

            state = next_state
            total_reward += reward
            steps += 1

        agent.decay_epsilon()
        episode_rewards.append(total_reward)
        episode_steps.append(steps)

        if (episode + 1) % 100 == 0:
            avg_reward = np.mean(episode_rewards[-100:])
            avg_steps = np.mean(episode_steps[-100:])
            print(f"Episode {episode + 1}/{episodes} - "
                  f"Avg Reward: {avg_reward:.2f}, "
                  f"Avg Steps: {avg_steps:.2f}, "
                  f"Epsilon: {agent.epsilon:.3f}")

    print("\n✓ Training Complete!\n")

    # Test the trained agent with visualization
    print("Testing Trained Agent (ALL Steps Visualization)...")
    state, _ = env.reset()
    done = False
    steps = 0
    agent_path = [state.copy()]

    # Store all positions
    all_positions = [state.copy()]

    while not done and steps < 200:
        action = agent.get_action(state)
        agent.epsilon = 0  # Use greedy policy for testing

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        agent_path.append(next_state.copy())
        all_positions.append(next_state.copy())
        state = next_state
        steps += 1

    print(f"Agent took {steps} steps to reach the goal!")
    print(f"Creating visualization for ALL {steps} steps...\n")

    # Calculate grid layout for all steps
    total_steps = len(all_positions)

    # Determine grid layout based on number of steps
    if total_steps <= 6:
        rows, cols = 2, 3
    elif total_steps <= 12:
        rows, cols = 3, 4
    elif total_steps <= 20:
        rows, cols = 4, 5
    elif total_steps <= 30:
        rows, cols = 5, 6
    else:
        rows, cols = 6, 8

    # Create multiple figures if needed
    steps_per_figure = rows * cols
    num_figures = (total_steps + steps_per_figure - 1) // steps_per_figure

    for fig_num in range(num_figures):
        start_idx = fig_num * steps_per_figure
        end_idx = min(start_idx + steps_per_figure, total_steps)
        steps_in_fig = end_idx - start_idx

        fig, axes = plt.subplots(rows, cols, figsize=(cols * 3, rows * 3))
        axes = axes.flatten()

        print(f"Generating figure {fig_num + 1}/{num_figures} (Steps {start_idx} to {end_idx - 1})...")

        for i in range(steps_per_figure):
            ax = axes[i]
            step_idx = start_idx + i

            if step_idx < total_steps:
                # Set agent position for this step
                env.agent_pos = all_positions[step_idx]

                # Get path up to this point
                current_path = all_positions[:step_idx + 1]

                # Render grid
                env.render_grid(ax, agent_path=current_path, current_step=step_idx)
            else:
                # Hide unused subplots
                ax.axis('off')

        plt.tight_layout()
        filename = f'gridworld_all_steps_part{fig_num + 1}.png'
        plt.savefig(filename, dpi=150, bbox_inches='tight')
        print(f"✓ Saved: {filename}")
        plt.show()

    # Plot training metrics
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot rewards
    ax1.plot(episode_rewards, alpha=0.3, color='blue')
    window = 50
    if len(episode_rewards) >= window:
        moving_avg = np.convolve(episode_rewards, np.ones(window)/window, mode='valid')
        ax1.plot(range(window-1, len(episode_rewards)), moving_avg,
                color='red', linewidth=2, label=f'{window}-episode moving avg')
    ax1.set_xlabel('Episode', fontsize=12)
    ax1.set_ylabel('Total Reward', fontsize=12)
    ax1.set_title('Training Rewards', fontsize=14, fontweight='bold')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot steps
    ax2.plot(episode_steps, alpha=0.3, color='green')
    if len(episode_steps) >= window:
        moving_avg = np.convolve(episode_steps, np.ones(window)/window, mode='valid')
        ax2.plot(range(window-1, len(episode_steps)), moving_avg,
                color='red', linewidth=2, label=f'{window}-episode moving avg')
    ax2.set_xlabel('Episode', fontsize=12)
    ax2.set_ylabel('Steps to Goal', fontsize=12)
    ax2.set_title('Training Steps', fontsize=14, fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('gridworld_qlearning_training.png', dpi=150, bbox_inches='tight')
    print(f"✓ Training metrics saved as 'gridworld_qlearning_training.png'")
    plt.show()

    print(f"\n{'='*60}")
    print(f"RESULTS:")
    print(f"{'='*60}")
    print(f"Goal reached: {done}")
    print(f"Total steps taken: {steps}")
    print(f"Final position: {state}")
    print(f"Target goal position: {env.goals[target_goal]}")
    print(f"Path length: {len(agent_path)}")
    print(f"All positions visited: {len(all_positions)}")
    print(f"{'='*60}\n")

    return env, agent, episode_rewards, episode_steps


# Run the training and visualization
if __name__ == "__main__":
    env, agent, rewards, steps = train_and_visualize(
        grid_size=4,
        target_goal=1,
        num_goals=2,
        num_obstacles=3,
        episodes=500
    )
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
from typing import Optional, Tuple, List
from collections import defaultdict

class MultiGoalGridWorld(gym.Env):
    metadata = {"render_modes": ["human", "rgb_array"], "render_fps": 4}

    def __init__(self, grid_size: int = 10, target_goal: int = 0,
                 num_goals: int = 3, num_obstacles: int = 10):
        super().__init__()

        self.grid_size = grid_size
        self.num_goals = num_goals
        self.target_goal = target_goal
        self.num_obstacles = num_obstacles

        # Define action and observation space
        # Actions: 0=UP, 1=RIGHT, 2=DOWN, 3=LEFT
        self.action_space = spaces.Discrete(4)

        # Observation: agent position (x, y)
        self.observation_space = spaces.Box(
            low=0, high=grid_size-1, shape=(2,), dtype=np.int32
        )

        # Initialize positions
        self.agent_pos = np.array([0, 0])
        self.goals = []
        self.obstacles = []
        self._generate_world()

    def _generate_world(self):
        """Generate random goal and obstacle positions"""
        self.goals = []
        self.obstacles = []

        # Get all available positions (excluding agent start)
        available_positions = [
            (x, y) for x in range(self.grid_size)
            for y in range(self.grid_size)
            if (x, y) != tuple(self.agent_pos)
        ]

        # Total positions needed
        total_needed = min(self.num_goals + self.num_obstacles, len(available_positions))

        # Randomly select positions
        selected_indices = np.random.choice(
            len(available_positions),
            size=total_needed,
            replace=False
        )

        # Assign goals
        for i in range(min(self.num_goals, total_needed)):
            idx = selected_indices[i]
            self.goals.append(np.array(available_positions[idx]))

        # Assign obstacles
        for i in range(self.num_goals, total_needed):
            idx = selected_indices[i]
            self.obstacles.append(np.array(available_positions[idx]))

    def _is_obstacle(self, pos: np.ndarray) -> bool:
        """Check if position contains an obstacle"""
        for obs in self.obstacles:
            if np.array_equal(pos, obs):
                return True
        return False

    def reset(self, seed: Optional[int] = None, options: Optional[dict] = None):
        super().reset(seed=seed)

        # Reset agent to fixed start position
        self.agent_pos = np.array([0, 0])

        # Ensure target_goal is valid
        if self.target_goal >= len(self.goals):
            self.target_goal = 0

        return self.agent_pos.copy(), {}

    def step(self, action: int):
        # Map actions to movements
        direction_map = {
            0: np.array([-1, 0]),  # UP
            1: np.array([0, 1]),   # RIGHT
            2: np.array([1, 0]),   # DOWN
            3: np.array([0, -1])   # LEFT
        }

        # Calculate new position
        new_pos = self.agent_pos + direction_map[action]

        # Check boundaries
        new_pos = np.clip(new_pos, 0, self.grid_size - 1)

        # Check if new position is an obstacle
        if not self._is_obstacle(new_pos):
            self.agent_pos = new_pos

        # Check if reached target goal
        target_pos = self.goals[self.target_goal]
        reached_goal = np.array_equal(self.agent_pos, target_pos)

        # Calculate reward
        if reached_goal:
            reward = 100.0
            terminated = True
        else:
            reward = -1.0  # Step penalty
            terminated = False

        truncated = False

        return self.agent_pos.copy(), reward, terminated, truncated, {}

    def render_grid(self, ax=None, agent_path=None, current_step=None):
        """Render the grid using matplotlib"""
        if ax is None:
            fig, ax = plt.subplots(figsize=(8, 8))

        ax.clear()
        ax.set_xlim(-0.5, self.grid_size - 0.5)
        ax.set_ylim(-0.5, self.grid_size - 0.5)
        ax.set_aspect('equal')
        ax.grid(True, alpha=0.3)
        ax.set_xticks(range(self.grid_size))
        ax.set_yticks(range(self.grid_size))
        ax.invert_yaxis()

        # Draw obstacles
        for obs in self.obstacles:
            rect = Rectangle((obs[1] - 0.5, obs[0] - 0.5), 1, 1,
                           facecolor='black', edgecolor='black')
            ax.add_patch(rect)

        # Draw goals
        for idx, goal in enumerate(self.goals):
            color = 'gold' if idx == self.target_goal else 'lightgray'
            circle = Circle((goal[1], goal[0]), 0.3, color=color, ec='black', linewidth=2)
            ax.add_patch(circle)
            ax.text(goal[1], goal[0], str(idx), ha='center', va='center',
                   fontsize=12, fontweight='bold')

        # Draw agent path if provided
        if agent_path and len(agent_path) > 1:
            path_array = np.array(agent_path)
            ax.plot(path_array[:, 1], path_array[:, 0], 'b--', alpha=0.5, linewidth=2, label='Path')

        # Draw agent
        agent_circle = Circle((self.agent_pos[1], self.agent_pos[0]), 0.3,
                             color='lime', ec='darkgreen', linewidth=2)
        ax.add_patch(agent_circle)

        title = f'GW - T Goal: {self.target_goal}'
        if current_step is not None:
            title += f' | Step: {current_step}'
        ax.set_title(title, fontsize=14, fontweight='bold')

        if agent_path and len(agent_path) > 1:
            ax.legend(loc='upper right')

        return ax


class QLearningAgent:
    """Q-Learning agent for GridWorld"""

    def __init__(self, action_space_size: int, learning_rate: float = 0.1,
                 discount_factor: float = 0.95, epsilon: float = 1.0,
                 epsilon_decay: float = 0.995, epsilon_min: float = 0.01):
        self.action_space_size = action_space_size
        self.lr = learning_rate
        self.gamma = discount_factor
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min

        # Q-table: defaultdict for dynamic state-action pairs
        self.q_table = defaultdict(lambda: np.zeros(action_space_size))

    def get_action(self, state: np.ndarray) -> int:
        """Choose action using epsilon-greedy policy"""
        state_key = tuple(state)

        if np.random.random() < self.epsilon:
            return np.random.randint(self.action_space_size)  # Explore
        else:
            return np.argmax(self.q_table[state_key])  # Exploit

    def update(self, state: np.ndarray, action: int, reward: float,
               next_state: np.ndarray, done: bool):
        """Update Q-table using Q-learning update rule"""
        state_key = tuple(state)
        next_state_key = tuple(next_state)

        # Q-learning update
        current_q = self.q_table[state_key][action]

        if done:
            target_q = reward
        else:
            max_next_q = np.max(self.q_table[next_state_key])
            target_q = reward + self.gamma * max_next_q

        # Update Q-value
        self.q_table[state_key][action] = current_q + self.lr * (target_q - current_q)

    def decay_epsilon(self):
        """Decay exploration rate"""
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)


def train_and_visualize(grid_size=10, num_goals=3, num_obstacles=15, episodes=500):
    """Train Q-learning agent with changing target goal per episode and visualize progress"""

    # Create environment
    env = MultiGoalGridWorld(
        grid_size=grid_size,
        target_goal=0,  # Initial target goal (will change per episode)
        num_goals=num_goals,
        num_obstacles=num_obstacles
    )

    # Create agent
    agent = QLearningAgent(
        action_space_size=env.action_space.n,
        learning_rate=0.1,
        discount_factor=0.95,
        epsilon=1.0,
        epsilon_decay=0.995,
        epsilon_min=0.01
    )

    # Training metrics
    episode_rewards = []
    episode_steps = []
    episode_goals = []  # Track target goals per episode

    print("Training Q-Learning Agent with Changing Target Goals...")
    print(f"Grid Size: {grid_size}x{grid_size}")
    print(f"Number of Goals: {num_goals}")
    print(f"Number of Obstacles: {num_obstacles}")
    print(f"Episodes: {episodes}\n")

    # Training loop
    for episode in range(episodes):
        # Set a new random target goal for this episode
        env.target_goal = np.random.randint(0, num_goals)
        episode_goals.append(env.target_goal)

        state, _ = env.reset()
        total_reward = 0
        steps = 0
        done = False

        while not done and steps < 200:  # Max steps per episode
            action = agent.get_action(state)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            agent.update(state, action, reward, next_state, done)

            state = next_state
            total_reward += reward
            steps += 1

        agent.decay_epsilon()
        episode_rewards.append(total_reward)
        episode_steps.append(steps)

        if (episode + 1) % 100 == 0:
            avg_reward = np.mean(episode_rewards[-100:])
            avg_steps = np.mean(episode_steps[-100:])
            most_common_goal = np.bincount(episode_goals[-100:]).argmax()
            print(f"Episode {episode + 1}/{episodes} - "
                  f"Avg Reward: {avg_reward:.2f}, "
                  f"Avg Steps: {avg_steps:.2f}, "
                  f"Most Common Goal: {most_common_goal}, "
                  f"Epsilon: {agent.epsilon:.3f}")

    print("\n✓ Training Complete!\n")

    # Test the trained agent with visualization (use final target goal)
    print("Testing Trained Agent (ALL Steps Visualization)...")
    env.target_goal = np.random.randint(0, num_goals)  # Random goal for testing
    print(f"Testing with Target Goal: {env.target_goal}")
    state, _ = env.reset()
    done = False
    steps = 0
    agent_path = [state.copy()]
    all_positions = [state.copy()]

    while not done and steps < 200:
        action = agent.get_action(state)
        agent.epsilon = 0  # Use greedy policy for testing

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        agent_path.append(next_state.copy())
        all_positions.append(next_state.copy())
        state = next_state
        steps += 1

    print(f"Agent took {steps} steps to reach the goal!")
    print(f"Creating visualization for ALL {steps} steps...\n")

    # Calculate grid layout for all steps
    total_steps = len(all_positions)

    # Determine grid layout based on number of steps
    if total_steps <= 6:
        rows, cols = 2, 3
    elif total_steps <= 12:
        rows, cols = 3, 4
    elif total_steps <= 20:
        rows, cols = 4, 5
    elif total_steps <= 30:
        rows, cols = 5, 6
    else:
        rows, cols = 6, 8

    # Create multiple figures if needed
    steps_per_figure = rows * cols
    num_figures = (total_steps + steps_per_figure - 1) // steps_per_figure

    for fig_num in range(num_figures):
        start_idx = fig_num * steps_per_figure
        end_idx = min(start_idx + steps_per_figure, total_steps)
        steps_in_fig = end_idx - start_idx

        fig, axes = plt.subplots(rows, cols, figsize=(cols * 3, rows * 3))
        axes = axes.flatten()

        print(f"Generating figure {fig_num + 1}/{num_figures} (Steps {start_idx} to {end_idx - 1})...")

        for i in range(steps_per_figure):
            ax = axes[i]
            step_idx = start_idx + i

            if step_idx < total_steps:
                env.agent_pos = all_positions[step_idx]
                current_path = all_positions[:step_idx + 1]
                env.render_grid(ax, agent_path=current_path, current_step=step_idx)
            else:
                ax.axis('off')

        plt.tight_layout()
        filename = f'gridworld_all_steps_part{fig_num + 1}.png'
        plt.savefig(filename, dpi=150, bbox_inches='tight')
        print(f"✓ Saved: {filename}")
        plt.show()

    # Plot training metrics
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot rewards
    ax1.plot(episode_rewards, alpha=0.3, color='blue')
    window = 50
    if len(episode_rewards) >= window:
        moving_avg = np.convolve(episode_rewards, np.ones(window)/window, mode='valid')
        ax1.plot(range(window-1, len(episode_rewards)), moving_avg,
                color='red', linewidth=2, label=f'{window}-episode moving avg')
    ax1.set_xlabel('Episode', fontsize=12)
    ax1.set_ylabel('Total Reward', fontsize=12)
    ax1.set_title('Training Rewards', fontsize=14, fontweight='bold')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot steps
    ax2.plot(episode_steps, alpha=0.3, color='green')
    if len(episode_steps) >= window:
        moving_avg = np.convolve(episode_steps, np.ones(window)/window, mode='valid')
        ax2.plot(range(window-1, len(episode_steps)), moving_avg,
                color='red', linewidth=2, label=f'{window}-episode moving avg')
    ax2.set_xlabel('Episode', fontsize=12)
    ax2.set_ylabel('Steps to Goal', fontsize=12)
    ax2.set_title('Training Steps', fontsize=14, fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('gridworld_qlearning_training.png', dpi=150, bbox_inches='tight')
    print(f"✓ Training metrics saved as 'gridworld_qlearning_training.png'")
    plt.show()

    print(f"\n{'='*60}")
    print(f"RESULTS:")
    print(f"{'='*60}")
    print(f"Goal reached: {done}")
    print(f"Total steps taken: {steps}")
    print(f"Final position: {state}")
    print(f"Target goal position: {env.goals[env.target_goal]}")
    print(f"Path length: {len(agent_path)}")
    print(f"All positions visited: {len(all_positions)}")
    print(f"{'='*60}\n")

    return env, agent, episode_rewards, episode_steps, episode_goals


# Run the training and visualization
if __name__ == "__main__":
    env, agent, rewards, steps, goals = train_and_visualize(
        grid_size=10,
        num_goals=4,
        num_obstacles=3,
        episodes=500
    )
```
